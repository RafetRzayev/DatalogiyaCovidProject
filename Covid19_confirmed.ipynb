```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('covid19_confirmed_global.csv')
data.head(10)
```

```python
data.describe()
```

```python
#We have a many null values in column 'Province/State', so let's drop it.
data.drop(columns ='Province/State', inplace = True)
```

```python
#We don't need Lat and Long columns, so let's drop it.
data.drop(columns =['Lat','Long'], inplace = True)
```

```python
data.head()
```

```python
last_week_data = data[["Country/Region","3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"]]
```

```python
last_week_data
```

```python
top_confirmed_countries = last_week_data.nlargest(n=10, columns=["3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"])
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('whitegrid')
top_confirmed_countries.plot(x="Country/Region", y=["3/24/22", "3/25/22", "3/26/22","3/27/22","3/28/22","3/29/22","3/30/22"], kind="bar")
```

```python
# Most confirmed cases happened in this countries when date is 3/30/22
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style('whitegrid')
sns.barplot(x='Country/Region',y='3/30/22',data=top_confirmed_countries)
```
